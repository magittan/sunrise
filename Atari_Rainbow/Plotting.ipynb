```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

# Asteroid Results

```python
fig = plt.figure(figsize=(10,10)) # create the canvas for plotting
ax1 = plt.subplot(2,2,1) 
ax2 = plt.subplot(2,2,2)
ax3 = plt.subplot(2,2,3)
ax4 = plt.subplot(2,2,4)

random_seeds = [123,231,312]
paths = ['results/sunrise_5/asteroids/Beta_0.5_T_10.0_UCB_I_0.0_UCB_T_0.0/seed_{}/metrics.pth'.format(seed) for seed in random_seeds]

colors = ['C0','C1','C2']

for i in range(3):
    results = torch.load(paths[i])
    steps = results['steps']
    average_rewards = np.array(results['rewards']).mean(axis=1)
    max_rewards = np.array(results['rewards']).max(axis=1)
    min_rewards = np.array(results['rewards']).min(axis=1)
    
    ax1.fill_between(steps, min_rewards, max_rewards, facecolor=colors[i], alpha=0.25)
    ax1.plot(steps, average_rewards)

ax1.set_ylim(100,1500)
ax1.set_ylabel("Score")
ax1.set_title("SUNRISE no UCB")

random_seeds = [123,231,312]
paths = ['results/sunrise_5/asteroids/Beta_0.5_T_10.0_UCB_I_1.0_UCB_T_0.0/seed_{}/metrics.pth'.format(seed) for seed in random_seeds]

colors = ['C0','C1','C2']

for i in range(3):
    results = torch.load(paths[i])
    steps = results['steps']
    average_rewards = np.array(results['rewards']).mean(axis=1)
    max_rewards = np.array(results['rewards']).max(axis=1)
    min_rewards = np.array(results['rewards']).min(axis=1)
    
    ax2.fill_between(steps, min_rewards, max_rewards, facecolor=colors[i], alpha=0.25)
    ax2.plot(steps, average_rewards)
    
ax2.set_ylim(100,1500)
ax2.set_title("SUNRISE")

random_seeds = [123,231,312]
paths = ['results/NHsunrise_5/asteroids/Beta_0.5_T_10.0_UCB_I_1.0_UCB_T_0.0/seed_{}/metrics.pth'.format(seed) for seed in random_seeds]

colors = ['C0','C1','C2']

for i in range(3):
    results = torch.load(paths[i])
    steps = results['steps']
    average_rewards = np.array(results['rewards']).mean(axis=1)
    max_rewards = np.array(results['rewards']).max(axis=1)
    min_rewards = np.array(results['rewards']).min(axis=1)
    
    ax3.fill_between(steps, min_rewards, max_rewards, facecolor=colors[i], alpha=0.25)
    ax3.plot(steps, average_rewards)

ax3.set_ylim(100,1500)
ax3.set_xlabel("Time Steps")
ax3.set_ylabel("Score")
ax3.set_title("R-SUNRISE")

random_seeds = [123,231,312]
paths = ['results/PCAsunrise_5/asteroids/Beta_0.5_T_10.0_UCB_I_1.0_UCB_T_0.0/seed_{}/metrics.pth'.format(seed) for seed in random_seeds]

colors = ['C0','C1','C2']

for i in range(3):
    results = torch.load(paths[i])
    steps = results['steps']
    average_rewards = np.array(results['rewards']).mean(axis=1)
    max_rewards = np.array(results['rewards']).max(axis=1)
    min_rewards = np.array(results['rewards']).min(axis=1)
    
    ax4.fill_between(steps, min_rewards, max_rewards, facecolor=colors[i], alpha=0.25)
    ax4.plot(steps, average_rewards)
ax4.set_title("U-SUNRISE")
ax4.set_ylim(100,1500)
ax4.set_xlabel("Time Steps")

plt.suptitle("Asteroid Training Plots over 3 Trials", y=0.925)
```
